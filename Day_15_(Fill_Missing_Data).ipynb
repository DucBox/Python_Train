```python
data = [1,2,3,None,6,7,3,None,5,None,5]
```

```python
for index, value in enumerate(data):
    if value is None:
        data[index] = data[index-1]
```

```python
for index, value in enumerate(data):
    if value is None:
      data[index] = (data[index-1]+data[index+1])/2
```
